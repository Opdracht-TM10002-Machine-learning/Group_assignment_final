```python
#script containing LDA algorithm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from worclipo.load_data import load_data
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from sklearn import decomposition
from sklearn import metrics
from sklearn import model_selection
from sklearn.metrics import RocCurveDisplay, accuracy_score
from scipy import stats
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.kernel_approximation import RBFSampler
from sklearn.metrics.pairwise import rbf_kernel, sigmoid_kernel
from scipy.stats import ttest_ind
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.feature_selection import RFECV
from sklearn.model_selection import LearningCurveDisplay, learning_curve
import seaborn as sns

data = load_data()
data.dropna(how='all')

print(f'The number of samples: {len(data.index)}')
print(f'The number of columns: {len(data.columns)}')

# Take the labels of the data
label = data.pop('label')
data = data.to_numpy()
label = label.to_numpy()

label1_aantal = [index for index, label in enumerate(label) if label == 'lipoma']
label2_aantal = [index for index, label in enumerate(label) if label == 'liposarcoma']

print(f"Het aantal labels in de klasse lipoma: {len(label1_aantal)}")
print(f"Het aantal labels in de klasse liposarcoma: {len(label2_aantal)}")

sss = model_selection.StratifiedKFold(n_splits=5)
mean_fpr = np.linspace(0, 1, 100)  # we maken een vector aan waarop we gaan interpoleren
tprs = []
auc = []
specificities = []
sensitivities = []

for train_index, test_index in sss.split(data, label):
    X_train = data[train_index]
    y_train = label[train_index]
    X_test = data[test_index]
    y_test = label[test_index]

    p_values_shapiro = []
    significant_p_values = []
    for column in range(X_train.shape[1]):
        _, p_value_shapiro = stats.shapiro(X_train[:,column])
        p_values_shapiro.append(p_value_shapiro)
    significant_p_values = sum(p < 0.05 for p in p_values_shapiro)
    print(f"Number of features that are not normally distributed {significant_p_values}")

    # Voer de t-test uit voor elke feature
    p_values = []

    label1_indices = [index for index, label in enumerate(y_train) if label == 'lipoma']
    label2_indices = [index for index, label in enumerate(y_train) if label == 'liposarcoma']

    for feature in range(X_train.shape[1]):  # num_features is het aantal features in je dataset
        group1 = X_train[label1_indices, feature]
        group2 = X_train[label2_indices, feature]
        _, p_value = ttest_ind(group1, group2)
        p_values.append(p_value)

    # Selecteer features met een p-waarde lager dan de drempel (bijv. 0.05)
    selected_features = [i for i, p_value in enumerate(p_values) if p_value < 0.02]

    # Drop de niet-significante features
    data_selected = X_train[:, selected_features]
    test_selected = X_test[:, selected_features]

    # Maak een DataFrame met de geselecteerde features
    selected_data = pd.DataFrame(data_selected)

    # Voeg de kolomnamen toe aan de DataFrame
    selected_data.columns = [f"feature_{i+1}" for i in selected_features]

    # Maken van een heatmap ter keuze PCA of RFE
    plt.figure(figsize=(12,8))
    sns.heatmap(selected_data.corr(), cmap="Blues")
    plt.show()

    correlations = []
    correlations = selected_data.corr()
    num_larger = (correlations > 0.3).sum().sum()
    total_sig = num_larger - len(correlations.columns)
    percentage_larger = total_sig / (correlations.size - len(correlations.columns))*100
    print(percentage_larger)

    # Definieer de pipeline met schaling, LDA en SVM-classifier
    pipeline = Pipeline([
        ('scaler', preprocessing.StandardScaler()),
        ('feature_selection', RFECV(estimator=SVC(kernel='linear'), scoring='accuracy')),
        ('lda', LinearDiscriminantAnalysis())
    ])

    # Definieer het parametergrid om te doorzoeken
    param_grid = {
        'scaler': [preprocessing.StandardScaler(), MinMaxScaler(), RobustScaler()],
        'feature_selection__step': [1, 5, 10],
        'lda__n_components': [None],
        'lda__solver': ['lsqr', 'eigen', 'svd'],  # De solver voor LDA
        'lda__shrinkage': ['auto'],  # L1 en L2 regularisatie (penalty)
    }

    # Initialiseer de grid search
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')

    # Voer de grid search uit op de data
    grid_search.fit(selected_data.values, y_train)

    # Bekijk de beste parameters en score
    print("Best parameters:", grid_search.best_params_)
    print("Validation accuracy:", grid_search.best_score_)

    # Voorspel op de testset met het beste model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(test_selected)
    y_proba = best_model.predict_proba(test_selected)[:, 1]

    #labels 0 of 1 maken
    label_encoder = preprocessing.LabelEncoder()
    y_test = label_encoder.fit_transform(y_test)
    y_pred = label_encoder.fit_transform(y_pred)

    # Evalueren van het beste model op de testset
    accuracy = accuracy_score(y_test, y_pred)
    print("Test set accuracy:", accuracy)

    train_size, train_score, test_score = learning_curve(best_model, selected_data, y_train)
    display = LearningCurveDisplay(train_sizes=train_size, train_scores=train_score, test_scores=test_score, score_name="Accuracy")
    display.plot()
    plt.show()


    confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
    sensitivity = confusion_matrix[1, 1] / (confusion_matrix[1, 1] + confusion_matrix[1, 0])
    specificity = confusion_matrix[0, 0] / (confusion_matrix[0, 0] + confusion_matrix[0, 1])

    auc_test = metrics.roc_auc_score(y_test, y_proba)
    fpr, tpr, threshold = metrics.roc_curve(y_test, y_proba)
    interp_tpr = np.interp(mean_fpr, fpr, tpr)
    interp_tpr[0] = 0

    tprs.append(interp_tpr)
    auc.append(auc_test)
    sensitivities.append(sensitivity)
    specificities.append(specificity)

mean_sens = np.mean(sensitivities)
mean_spec = np.mean(specificities)
print(f"The mean sensitivity of the model is {mean_sens}")
print(f"The mean specificity of the model is {mean_spec}")

tpr_mean = np.mean(tprs, axis=0)
tpr_mean[-1] = 1.0
auc_mean = metrics.auc(mean_fpr, tpr_mean)
std_auc = np.std(auc)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(
    mean_fpr,
    tpr_mean,
    color="b",
    label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (auc_mean, std_auc),
    lw=2,
    alpha=0.8,
)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(tpr_mean + std_tpr, 1)
tprs_lower = np.maximum(tpr_mean - std_tpr, 0)
ax.fill_between(
    mean_fpr,
    tprs_lower,
    tprs_upper,
    color="grey",
    alpha=0.2,
    label=r"$\pm$ 1 std. dev.",
)

ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title=f"Mean ROC curve with variability\n(Positive label '{label_encoder.classes_[1]}')",
)
ax.legend(loc="lower right")
plt.show()
```

```python
#script containing RF
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from worclipo.load_data import load_data
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing
from sklearn import metrics
from sklearn import model_selection
from sklearn.metrics import RocCurveDisplay, accuracy_score
from scipy import stats
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from scipy.stats import ttest_ind
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import ttest_ind
from sklearn.feature_selection import RFECV

data = load_data()
data.dropna(how='all')

print(f'The number of samples: {len(data.index)}')
print(f'The number of columns: {len(data.columns)}')

# Take the labels of the data
label = data.pop('label')
data = data.to_numpy()
label = label.to_numpy()

sss = model_selection.StratifiedKFold(n_splits=5)
mean_fpr = np.linspace(0, 1, 100)  # we maken een vector aan waarop we gaan interpoleren
tprs = []
auc = []
specificities = []
sensitivities = []

for train_index, test_index in sss.split(data, label):
    X_train = data[train_index]
    y_train = label[train_index]
    X_test = data[test_index]
    y_test = label[test_index]

    # Voer de t-test uit voor elke feature
    p_values = []

    label1_indices = [index for index, label in enumerate(y_train) if label == 'lipoma']
    label2_indices = [index for index, label in enumerate(y_train) if label == 'liposarcoma']

    for feature in range(X_train.shape[1]):  # num_features is het aantal features in je dataset
        group1 = X_train[label1_indices, feature]
        group2 = X_train[label2_indices, feature]
        _, p_value = ttest_ind(group1, group2)
        p_values.append(p_value)

    # Selecteer features met een p-waarde lager dan de drempel (bijv. 0.05)
    selected_features = [i for i, p_value in enumerate(p_values) if p_value < 0.05]

    # Drop de niet-significante features
    data_selected = X_train[:, selected_features]
    test_selected = X_test[:, selected_features]

    # Maak een DataFrame met de geselecteerde features
    selected_data = pd.DataFrame(data_selected)

    # Voeg de kolomnamen toe aan de DataFrame
    selected_data.columns = [f"feature_{i+1}" for i in selected_features]

    # Definieer de pipeline met schaling en Random Forest classifier
    pipeline = Pipeline([
        ('scaler', preprocessing.StandardScaler()),
        ('feature_selection', RFECV(estimator=SVC(kernel='linear'), scoring='accuracy')),  # Voeg RFECV toe als feature selector
        ('rf', RandomForestClassifier())
    ])

    # Definieer het parametergrid om te doorzoeken
    param_grid = {
        'scaler':  [MinMaxScaler(), RobustScaler(), StandardScaler()],
        'feature_selection__step': [1, 5, 10],  # Aantal features om per iteratie te verwijderen
        'rf__n_estimators': [2, 10, 40, 80],  # Aantal bomen in het forest
        'rf__max_depth': [2, 4, 6, 8, 10],  # Maximale diepte van de bomen
    }

    # Initialiseer de grid search
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')

    # Voer de grid search uit op de data
    grid_search.fit(selected_data, y_train)

    # Bekijk de beste parameters en score
    print("Best parameters:", grid_search.best_params_)
    print("Best accuracy:", grid_search.best_score_)

    # Voorspel op de testset met het beste model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(test_selected)
    y_proba = best_model.predict_proba(test_selected)[:, 1]

    #labels 0 of 1 maken
    label_encoder = preprocessing.LabelEncoder()
    y_test = label_encoder.fit_transform(y_test)
    y_pred = label_encoder.fit_transform(y_pred)

    # Evalueren van het beste model op de testset
    accuracy = accuracy_score(y_test, y_pred)
    print("Test set accuracy:", accuracy)

    train_size, train_score, test_score = learning_curve(best_model, selected_data, y_train)
    display = LearningCurveDisplay(train_sizes=train_size, train_scores=train_score, test_scores=test_score, score_name="Accuracy")
    display.plot()
    plt.show(block=False)

    confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
    sensitivity = confusion_matrix[1, 1] / (confusion_matrix[1, 1] + confusion_matrix[1, 0])
    specificity = confusion_matrix[0, 0] / (confusion_matrix[0, 0] + confusion_matrix[0, 1])

    auc_test = metrics.roc_auc_score(y_test, y_proba)
    fpr, tpr, threshold = metrics.roc_curve(y_test, y_proba)
    interp_tpr = np.interp(mean_fpr, fpr, tpr)
    interp_tpr[0] = 0

    tprs.append(interp_tpr)
    auc.append(auc_test)
    sensitivities.append(sensitivity)
    specificities.append(specificity)

mean_sens = np.mean(sensitivities)
mean_spec = np.mean(specificities)
print(f"The mean sensitivity of the model is {mean_sens}")
print(f"The mean specificity of the model is {mean_spec}")

tpr_mean = np.mean(tprs, axis=0)
tpr_mean[-1] = 1.0
auc_mean = metrics.auc(mean_fpr, tpr_mean)
std_auc = np.std(auc)

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(
    mean_fpr,
    tpr_mean,
    color="b",
    label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (auc_mean, std_auc),
    lw=2,
    alpha=0.8,
)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(tpr_mean + std_tpr, 1)
tprs_lower = np.maximum(tpr_mean - std_tpr, 0)
ax.fill_between(
    mean_fpr,
    tprs_lower,
    tprs_upper,
    color="grey",
    alpha=0.2,
    label=r"$\pm$ 1 std. dev.",
)

ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title=f"Mean ROC curve with variability\n(Positive label '{label_encoder.classes_[1]}')",
)
ax.legend(loc="lower right")
plt.show()
```
